Campaign Performance:

```python
# import numpy and pandas libraries
import numpy as np
import pandas as pd
```

```python
# a) Plot a scatter plot to visualize the relationship between Acquisition_Cost and ROI.
```

```python
plt.figure(figsize = (10,5))
plt.scatter(df_csv['Acquisition_Cost'], df_csv['ROI'], color = 'Grey', alpha = 0.75)
plt.title('Scatter Plot: Acquisition Cost vs ROI')
plt.xlabel('Acquisition Cost')
plt.ylabel('ROI')
plt.grid(True)
plt.show()
```

```python
# b) Create a bar chart to visualize the average Conversion_Rate for different Channel_Used, categorized by Campaign_Type.
```

```python
ax = df_csv.groupby(['Channel_Used','Campaign_Type'])['Conversion_Rate'].mean().plot(kind= 'bar', figsize =(20,8))
ax.bar_label(ax.containers[0])
plt.title('Average Conversion rate for different Channel used & Categorized by Campaign Type')
plt.xlabel('Channel_Used')
plt.ylabel('Conversion Rate')
plt.grid(True)
plt.show()
```

```python
# c) Visualize the distribution of Engagement_Score across different Campaign_Type using a box plot.
```

```python
plt.figure(figsize = (15,10))
df_csv.boxplot(column =['Engagement_Score'], by = 'Campaign_Type')
plt.title('Distribution of Engagement Score across different Campaign type ')
plt.xlabel('Campaign Type')
plt.ylabel('Engagement Score')
plt.grid(True)
plt.show()
```

```python
# d) Analyze the average ROI by Company using a bar chart to compare the profitability of campaigns conducted by different companies.
```

```python
ax = df_csv.groupby('Company')['ROI'].mean().plot(kind= 'bar', figsize = (10,5))
ax.bar_label(ax.containers[0])
plt.title('Average ROI to compare the profitability of Campaigns conducted by different Companies')
plt.xlabel('Company')
plt.ylabel('ROI')
plt.grid(True)
plt.xticks(rotation = 45 )
plt.show()
```

```python
# e) âž¢ Examine the correlation between Engagement_Score and Conversion_Rate using a heatmap.
```

```python
import seaborn as sns
```

```python
correlation_matrix = df_csv[['Engagement_Score', 'Conversion_Rate']].corr()
sns.heatmap(correlation_matrix,annot= True)
```

 Customer Segmentation:

```python
# a) âž¢ Create a count plot to visualize the distribution of Target_Audience.
```

```python
sns.countplot(x='Target_Audience', data=df_csv, palette='Set2')
plt.title("Count of Distribution of Target Audience")
plt.show()
```

```python
# b) Identify which Customer_Segment has the highest Conversion_Rate for each Language using a bar chart.
```

```python
ax = df_csv.groupby(['Customer_Segment','Language'])['Conversion_Rate'].max().plot(kind= 'bar', figsize =(20,8), color = 'pink')
ax.bar_label(ax.containers[0])
plt.title('Highest conversion rate for each language using a bar chart')
plt.xlabel('Customer Segment')
plt.ylabel('Conversion Rate')
plt.grid(True)
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# c) Visualize the distribution of Acquisition_Cost across each Customer_Segment, categorized by Channel_Used, using a box plot.
```

```python
plt.figure(figsize = (15,10))
df_csv.boxplot(column =['Acquisition_Cost'], by = 'Channel_Used')
plt.title('Distribution of Acquisition Cost Across each Customer Segment categorized by Channel used')
plt.xlabel('Channel used')
plt.ylabel('Acquistion Cost')
plt.grid(True)
plt.show()
```

```python
# d) âž¢ Analyze average Conversion_Rate by Language using a bar chart to compare the effectiveness of campaigns conducted in different languages.
```

```python
ax = df_csv.groupby('Language')['Conversion_Rate'].mean().plot(kind= 'bar', figsize =(6,3))
Average_Conversion_Rate = df_csv.groupby('Language')['Conversion_Rate'].mean().reset_index()
ax = plt.bar(Average_Conversion_Rate['Language'], Average_Conversion_Rate['Conversion_Rate'], color='Turquoise')
plt.title('Effectiveness of Campaign conducted in different languages')
plt.xlabel('Languages')
plt.ylabel('Conversion Rate')
plt.grid(False)
plt.xticks(rotation=45, ha='right')
plt.show()
```

Channel Effectiveness:

```python
# a) Compare the Engagement_Score for different Channels_Used, segmented by Campaign_Type, using a bar chart.
```

```python
ax = df_csv.groupby(['Channel_Used','Campaign_Type'])['Engagement_Score'].mean().plot(kind= 'bar', stacked = True, figsize =(20,8))
ax.bar_label(ax.containers[0])
plt.title('Comparison of Engagement scores for different channels used segmented by campaign type')
plt.xlabel('Channel_Used')
plt.ylabel('Engagement_Score')
plt.grid(True)
plt.show()
```

```python
# b) Show the distribution of total ROI across different Channels_Used using a pie chart.
```

```python
ROI_by_Channel = df_csv.groupby('Channel_Used')['ROI'].sum()
plt.pie(ROI_by_Channel,labels=ROI_by_Channel.index, autopct='%1.3f%%', startangle = 108)
plt.title('Distribution of Total ROI across different channels used')
plt.show()
```

```python
# c) Plot a scatter plot to show the relationship between Clicks and Impressions for each Channel_Used.
```

```python
plt.figure(figsize = (10,5))
plt.scatter(df_csv['Clicks'], df_csv['Impressions'], color = 'brown', alpha = 0.90, marker = 'x')
plt.title('Scatter plot: Relationship between clicks and Impressions for each channel used')
plt.xlabel('Clicks')
plt.ylabel('Impressions')
plt.grid(True)
plt.show()
```

Time-Based Analysis:

a) Plot the distribution of Duration using a histogram.

```python
plt.figure(figsize=(7, 5))
plt.hist(df_csv['Duration'], bins=20, color='Yellow', edgecolor='black')
plt.title('Distribution of Duration')
plt.xlabel('Duration')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

b) Analyze how the overall Conversion_Rate has changed over Date for each Company using a line chart.

```python
df_csv['Date'] = pd.to_datetime(df_csv['Date'], format='%d-%m-%Y')
Company_CR = df_csv.groupby(['Date', 'Company'])['Conversion_Rate'].mean().reset_index()
for company in Company_CR['Company'].unique():
    company_data = Company_CR[Company_CR['Company'] == company]
plt.plot(company_data['Date'], company_data['Conversion_Rate'], label=company)
plt.title('Overall Conversion rate changed over date')
plt.xlabel('Date')
plt.ylabel('Avg Conversion date')
plt.legend(title= 'Company')
plt.grid(True)
plt.show()

```

c) Examine the trend of Engagement_Score over Date with a line chart

```python
df_csv['Date'] = pd.to_datetime(df_csv['Date'], format='%d-%m-%Y')
plt.plot(df_csv['Date'], df_csv['Engagement_Score'])
plt.title('Engagement Score over date with a line chart')
plt.xlabel('Date')
plt.ylabel('Engagement Score')
plt.grid(True)
plt.show()
```

Geographic Analysis:

âž¢ Visualize the Conversion_Rate by different Location, categorized by Target_Audience, using a bar chart.

```python
ax = df_csv.groupby(['Location','Target_Audience'])['Conversion_Rate'].mean().plot(kind= 'bar', figsize =(20,8))
ax.bar_label(ax.containers[0])
plt.title('Conversion rate by different location categorized by Target audience')
plt.xlabel('Location')
plt.ylabel('Average Conversion rate')
plt.grid(True)
plt.xticks(rotation=45, ha='right')
plt.show()
```
